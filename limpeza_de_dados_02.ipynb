```python
# BIBLIOTECA PANDAS
import pandas as pd
```

```python
# CARREGAR A BASE DE DADOS
dados = pd.read_csv('/content/drive/MyDrive/athlete_events.csv')
# ABRIR 
dados.head()
```

```python
# VER O TOTAL DE LINHAS E COLUNAS
dados.shape
```

```python
# CRIAR UMA NOVA VARIÁVEL E EXCLUIR TODAS AS LINHAS QUE TENHAM DADOS VAZIOS 
# NEM SEMPRE ESSE MÉTODO É UTIL, PRINCIPALMENTE SI A QUANTIDADE DE DADOS EXCLUIDAS FOREM MUITO GRANDES OU DE MUITA IMPORTÂNCIA
dados2 = dados.dropna()
dados2.head()
```

```python
# MOSTRAR O TOTAL DE LINHAS E COLUNAS QUE FICOU APÓS EXCLUIR OS DADOS VAZIOS DO DATASET
# OBS: NESSE CASO NÃO É UMA BOA ESTRATÉGIA EXCLUIR PORQUE OUVE MUITOS DADOS PERDIDOS DA BASE DE DADOS
dados2.shape
```

```python
# USAR A FUNÇÃO isnull PARA VER ONDE TEM DADOS FALTANTE
# OBS: TUDO QUE TIVER COM O NOME DE 'TRUE' É DADOS FALTANTES
enulo = dados.isnull()
enulo.head()
```

```python
# CRIO UMA NOVA VARIÁVEL E APLICO A FUNÇÃO PARA SOMA O TOTAL DE DADOS VAZIOS DE CADA COLUNA
faltantes = dados.isnull().sum()
print(faltantes)
```

```python
# MOSTRAR EM PORCENTAGEM COMO FICOU A REPRESENTAÇÃO DOS DADOS ACIMA
faltantes_porcentagem = (dados.isnull().sum() / len(dados['ID']))
print(faltantes_porcentagem)
```

## TRATAMENTO DE DADOS

```python
# NA COLUNA MEDALHA SUBSTITUI NaN POR nenhum
dados['Medal'].fillna('nenhuma', inplace=True)
# NA COLUNA IDADE VOU FAZER O CALCULO DA MÉDIA
dados['Age'].fillna(dados['Age'].mean(), inplace=True)
# NA COLUNA ALTURA VOU VER TAMBÉM O CALCULO DA MÉDIA
dados['Height'].fillna(dados['Height'].mean(), inplace=True)
# NA COLUNA PESO TAMBÉM FAZER A MÉDIA
dados['Weight'].fillna(dados['Weight'].mean(), inplace=True)
# MOSTRAR AS 100 PRIMEIRAS LINHAS
dados.head(100)
```

```python
# MANTIVE TODAS AS LINHAS E TODAS AS COLUNAS SEM PRECISAR EXCLUIR COLUNAS
dados.shape
```

```python
# MOSTRAR NOVAMENTE A TABELA DE VALORES VAZIOS SENDO QUE AGORA NÃO TEMOS MAIS DADOS FALTANTES
faltantes_porcentagem = (dados.isnull().sum() / len(dados['ID']))
print(faltantes_porcentagem)
```
